&nbsp;&nbsp;&nbsp;&nbsp;Логистическая регрессия или логит-модель (англ. logit model) - статистическая модель, используемая для прогнозирования вероятности возникновения некоторого события путём его сравнения с логистической кривой. Эта регрессия выдаёт ответ в виде вероятности бинарного события (1 или 0). Можно называть модификацией линейной регрессии, переходим от предсказания непрерывной случайной величины в дискретную случайную величину при помощи логистической функции (сигмоиды).

&nbsp;&nbsp;&nbsp;&nbsp;Для этого вводится так называемая ***зависимая переменная y<sub>i</sub>***, принимающая лишь одно из двух значений - как правило, это числа 0 (событие не произошло) и 1 (событие произошло), ***и множество независимых переменных*** (также называемых признаками, предикторами или регрессорами) - вещественных ***x<sub>1</sub>,x<sub>2</sub>, ⋯, x<sub>n</sub>***.

$$ z=θ_0+θ_1​x_1​+θ_2​x_2​+⋯+θ_n​x_n $$

&nbsp;&nbsp;&nbsp;&nbsp;Прогоняем через сигмоиду:

$$ σ(z)=\frac{1}{1 + e^{-z}} $$

&nbsp;&nbsp;&nbsp;&nbsp;Это гарантирует, что предсказанные значения лежат в диапазоне от 0 до 1.

**Подбор параметров.**
Для подбора параметров *θ<sub>0</sub>, θ<sub>1</sub>, ⋯, θ<sub>n</sub>* необходимо составить обучающую выборку, состоящую из наборов значений независимых переменных и соответствующих им значений зависимой переменной y. Формально, это множество пар (x, y), где x - вектор независимых переменных примера, а y - зависимая переменная примера (наблюдаемая) в диапозоне от 0 до 1. Вместе они - обучающий пример.

&nbsp;&nbsp;&nbsp;&nbsp;ROC-кривая (англ. receiver operating characteristic, рабочая характеристика приёмника) - график, позволяющий оценить качество бинарной классификации, отображает соотношение между долей объектов от общего количества носителей признака, верно классифицированных как несущие признак (англ. true positive rate, TPR, называемой чувствительностью алгоритма классификации), и долей объектов от общего количества объектов, не несущих признака, ошибочно классифицированных как несущие признак (англ. false positive rate, FPR, величина 1-FPR называется специфичностью алгоритма классификации) при варьировании порога решающего правила.

&nbsp;&nbsp;&nbsp;&nbsp;Также известна как кривая ошибок. Анализ классификаций с применением ROC-кривых называется ROC-анализом.

&nbsp;&nbsp;&nbsp;&nbsp;Количественная интерпретация ROC даёт показатель AUC (англ. Area Under Curve, площадь под кривой) - площадь, ограниченная ROC-кривой и осью доли ложных положительных классификаций. Чем выше показатель AUC, тем качественнее классификатор, при этом значение 0,5 демонстрирует непригодность выбранного метода классификации (соответствует случайному гаданию). Значение менее 0,5 говорит, что классификатор действует с точностью до наоборот: если положительные назвать отрицательными и наоборот, классификатор будет работать

&nbsp;&nbsp;&nbsp;&nbsp;Задача классификации состоит в том, чтобы относить ранее неизвестные объекты к тому или иному классу. Примером такой задачи может быть диагностика болезни — заболел пациент (положительный результат) или нет (отрицательный результат). Тогда в результате классификации может наблюдаться четыре различных ситуации:
* истинно-положительный результат (true-positive, TP) — пациент больной, диагноз положительный;
* ложно-положительный результат (false-positive, FP) — пациент здоров, диагноз положительный;
* истинно-отрицательный результат (true-negative, TN) — пациент здоров, диагноз отрицательный;
* ложно-отрицательный результат (false-negative, FN) — пациент больной, диагноз отрицательный.

&nbsp;&nbsp;&nbsp;&nbsp;Порог - минимальное предсказанное значение, от которого будем считать принадлежность к классу.


  
 

***Идеальная модель:*** 

&nbsp;&nbsp;&nbsp;&nbsp;ROC-кривая проходит через точку (0, 1), где 𝐹𝑃𝑅 = 0, 𝑇𝑃𝑅 = 1, что соответствует AUC = 1. Это значит, что модель всегда верно классифицирует оба класса. 

***Cлучайная модель:***

&nbsp;&nbsp;&nbsp;&nbsp;ROC-кривая будет диагональю (0, 0) → (1, 1), где AUC = 0.5. Это указывает на отсутствие способности модели различать классы. 

***Плохая модель:*** 

&nbsp;&nbsp;&nbsp;&nbsp;Если AUC < 0.5, модель хуже случайной и может быть инвертирована для улучшения результата.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
display(
    train,
    test,
    train.info(),
    test.info(),
)
```

```python
train.dropna(subset=['Financial Stress', 'Degree', 'Dietary Habits'])

median_job_satis = train['Job Satisfaction'].median()
train['Job Satisfaction'] = train['Job Satisfaction'].fillna(median_job_satis)

median_work_pressure = train['Work Pressure'].median()
train['Work Pressure'] = train['Work Pressure'].fillna(median_work_pressure)
```

***Подготовка категориальных признаков***

```python
display(train.dtypes)

train['Working Professional or Student'] = train['Working Professional or Student'].replace({'Working Professional': 1, 'Student': 0, 'nan': 0}).astype(int)

freq = train['Sleep Duration'].value_counts(normalize=True)
train['Sleep Duration'] = train['Sleep Duration'].map(freq)

train['Have you ever had suicidal thoughts ?'] = train['Have you ever had suicidal thoughts ?'].replace({'Yes': 1, 'No': 0, 'nan': 0}).astype(int)

train['Family History of Mental Illness'] = train['Family History of Mental Illness'].replace({'Yes': 1, 'No': 0, 'nan': 0}).astype(int)

train = pd.get_dummies(train, columns=['Gender'], dtype=float)
#train.drop(['Age'])

display(train)
```

```python
X_test = test[['Age', 'Work/Study Hours', 'Financial Stress']]
X = train[['Age', 'Work/Study Hours', 'Financial Stress']]

X['Financial Stress'].fillna(0, inplace=True) 
X_test['Financial Stress'].fillna(0, inplace=True)

y = train['Depression']

X_train_, X_valid_, y_train_, y_valid_ = train_test_split(X, y, test_size=0.3, random_state=42)

display(
    X_train_.shape,
    y_train_.shape,
    X_valid_.shape,
    y_valid_.shape,
    y_train_.value_counts(True),
    y_valid_.value_counts(True),
)
```

```python
# Обучаем модель
model = LogisticRegression(random_state=42, max_iter=1000)
# model = RandomForestClassifier(random_state=42)
model.fit(X_train_, y_train_)
```

```python
y_pred = model.predict(X_valid_)
```

```python
# Оценка
y_pred_proba = model.predict_proba(X_valid_)
roc_auc = roc_auc_score(y_valid_, y_pred_proba[:, 1])

print(f'ROC AUC Score: {roc_auc}')

# Вычисление ROC-метрик
fpr, tpr, thresholds = roc_curve(y_valid_, y_pred_proba[:, 1])  # FPR, TPR и пороги
roc_auc = auc(fpr, tpr)  # Площадь под кривой (AUC)

# Построение графика
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')  # ROC-кривая
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random model')  # Линия случайной модели
plt.xlim([0.0, 1.0])  # Ограничения оси X
plt.ylim([0.0, 1.05])  # Ограничения оси Y
plt.xlabel('False Positive Rate')  # Подпись оси X
plt.ylabel('True Positive Rate')  # Подпись оси Y
plt.title('Receiver Operating Characteristic (ROC)')  # Заголовок
plt.legend(loc="lower right")  # Легенда
plt.show()

y_pred_proba_test = model.predict_proba(X_test)
pd.DataFrame(y_pred_proba_test, columns=['Healty', 'Depression']).to_csv('submission.csv', index=False)
```
