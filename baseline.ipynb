```python
import pandas as pd
import numpy as np
import matplotlib as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')

X_test = test[['Work/Study Hours', 'Financial Stress']]
X = train[['Work/Study Hours', 'Financial Stress']]

X['Financial Stress'].fillna(0, inplace=True) 
X_test['Financial Stress'].fillna(0, inplace=True)

y = train['Depression']

display(
    train.shape,
    test.shape,
    train.
)
```

```python
X_train_, X_valid_, y_train_, y_valid_ = train_test_split(X, y, test_size=0.3, random_state=42)

display(
    X_train_.shape,
    y_train_.shape,
    X_valid_.shape,
    y_valid_.shape,
    y_train_.value_counts(True),
    y_valid_.value_counts(True),
)
```

```python
#  
model = LogisticRegression(random_state=42, max_iter=1000)
# model = RandomForestClassifier(random_state=42)
model.fit(X_train_, y_train_)
```

```python
y_pred = model.predict(X_valid_)
```

```python
# 
y_pred_proba = model.predict_proba(X_valid_)
roc_auc = roc_auc_score(y_valid_, y_pred_proba[:, 1])

print(f'ROC AUC Score: {roc_auc}')

#  ROC-
fpr, tpr, thresholds = roc_curve(y_valid_, y_pred_proba[:, 1])  # FPR, TPR  
roc_auc = auc(fpr, tpr)  #    (AUC)

#  
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')  # ROC-
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random model')  #   
plt.xlim([0.0, 1.0])  #   X
plt.ylim([0.0, 1.05])  #   Y
plt.xlabel('False Positive Rate')  #   X
plt.ylabel('True Positive Rate')  #   Y
plt.title('Receiver Operating Characteristic (ROC)')  # 
plt.legend(loc="lower right")  # 
plt.show()

y_pred_proba_test = model.predict_proba(X_test)
pd.DataFrame(y_pred_proba_test, columns=['Healty', 'Depression']).to_csv('submission.csv', index=False)
```
